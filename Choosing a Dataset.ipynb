Used this [guide](https://www.geeksforgeeks.org/how-to-list-tables-using-sqlite3-in-python/) to get the list of tables.

```python
# Adatped from https://www.geeksforgeeks.org/how-to-list-tables-using-sqlite3-in-python/

cursor =  connection.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
cursor.close()
```
